```python
import numpy as np
import networkx as nx
import random
```

```python
vertex=['A','B','C','D']
edges=[('A','B'),('A','C'),('A','D'),('B','A'),('B','C'),('C','A'),('C','D'),('B','D')]
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from(vertex)
G.add_edges_from(edges)
```

```python
nx.draw(G,with_labels=True,node_color='red',node_size=1000)
```

```python
def randomwalk(Graph,node,node2,limit):
    rand_walk=[node]
    while(node!=node2 and limit-1):
        temp=list(Graph.neighbors(node))
        temp=list(set(temp)-set(rand_walk))
        if temp==0:
            break
        new_walk=np.random.choice(temp)
        rand_walk.append(new_walk)
        node=new_walk
        limit-=1
    return rand_walk
```

```python
print(randomwalk(G,'D','B',3))
```
